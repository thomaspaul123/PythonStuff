```python

import os
import matplotlib.pyplot as plt 
from matplotlib.figure import Figure
from PIL import Image
import shutil

def data_nav_lab(source, destination):
    
    # create a folder for each label
    labels=['Cracks', 'Groove', 'Rut', 'Subsidence']
    for f in labels:
        try:
            os.mkdir(os.path.join(destination, f))
        except:
            pass
        
    # show image and label it manually
    for photo in os.listdir(source):
        plt.figure(figsize=(15, 8))        
        plt.imshow(Image.open(os.path.join(source, photo)))
        plt.show()
        # choose the label
        name = input('Pass:0, Cracks:1, Groove:2, Rut:3, Subsidence:4')
        if len(name.split())==1:
            if name==0:
                continue
            typ = labels[int(name)-1]
            leng = len(os.listdir(os.path.join(destination ,typ)))
            shutil.move(os.path.join(source ,photo),
                        os.path.join(destination ,typ, photo+str(leng)+'.jpg'))
        
        # if image belonges to multiple labels
        else:
            for n in name.split():
                typ = labels[int(n)-1]
                leng = len(os.listdir('Data Omdena/{}'.format(typ)))
                shutil.copy(os.path.join(source ,photo),
                            os.path.join(destination ,typ, photo+str(leng)+'.jpg'))
            os.remove(os.path.join(source ,photo))
            
data_nav_lab(r'C:\My_work\Python\nimmi\source', r'C:\My_work\Python\nimmi\dest')
```
